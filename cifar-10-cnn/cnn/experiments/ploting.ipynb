```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path = 'architecture/1.0.0/'
csv_name = 'training.csv'
fig_name = 'v.1.0.0.png'
csv = np.genfromtxt(path + csv_name, delimiter=',')
```

```python
# Using Numpy to create an array X
# X = np.arange(0, math.pi*2, 0.05)
  
# Naming the x-axis, y-axis and the whole graph
# plt.xlabel("Angle")
# plt.ylabel("Magnitude")
# plt.title("Sine and Cosine functions")
  
# Adding legend, which helps us recognize the curve according to it's color
# plt.legend()
  
fig, axs = plt.subplots(2, figsize=(16, 8))
fig.suptitle('architecture v1.0.0')
axs[0].set_title('loss')
axs[0].plot(csv[:,0], color='r', label='trainset')
axs[0].plot(csv[:,2], color='g', label='testset')
axs[1].set_title('accuracy')
axs[1].plot(csv[:,1], color='r', label='trainset')
axs[1].plot(csv[:,3], color='g', label='testset')
```

```python
# fig.savefig(path + fig_name)

# Save just the portion _inside_ the second axis's boundaries
extent = axs[1].get_window_extent().transformed(fig.dpi_scale_trans.inverted())
fig.savefig(path + 'figure.png', bbox_inches=extent)

# Pad the saved area by 10% in the x-direction and 20% in the y-direction
fig.savefig(path + 'figure_expanded.png', bbox_inches=extent.expanded(1.1, 1.2))
```
